```python
import numpy as np
import random
import numpy.random as npr
```

```python
npr.seed(0)  #  generar una varibale aleatoria uniforme (0,1)
npr.rand()
```

```python
muestra =npr.random(size=10)
muestra
```

```python
muestra_uniformes = npr.uniform(0,2,10)
muestra_uniformes
```

```python

npr.seed(1)
muestras_binomiales = npr.binomial(1,0.6, 1000)
muestras_binomiales
```

```python
import pandas as pd
df = pd.DataFrame()

for i in range (1,100):
  muestras_binomiales = npr.binomial(1000, 0.6, 50)
  col = f"muestras_{i}"
  df [col] = muestras_binomiales

df_media = pd.DataFrame(  df.mean(), columns=["muestra de las medias "])
df_media
```

```python
import seaborn as sns

sns.distplot(df_media)
```

```python
import seaborn as sns

df = pd.DataFrame()

for i in range (1,100):
  muestras_exponencial = npr.exponential(1,  50)
  col = f"muestras_{i}"
  df [col] = muestras_exponencial

df_media = pd.DataFrame(  df.mean(), columns=["muestra de las medias "])
sns.distplot(df_media)
```
